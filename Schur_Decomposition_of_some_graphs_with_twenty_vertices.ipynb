```python
#Partitions of nine.
Partition_of_nine=list(Partitions(9)) 
#We set by Some_graph_of_twenty_vertices the list of all graphs with twenty vertices
#such that in those graphs, there is a vertex v0 with nine neighborhoord and the other vertices is either 
#a neighborhood of v0 or neighborhoord of its neigborhoord.
Some_graph_of_twenty_vertices=[] #Initialization.
#We initialize all graphs with a dictionary which contains v0 with ten neighborhoord v1,..,v10.
Graph_Initialization={0:list(range(1,11))} 
vertex=1
start=11
for v in Partition_of_nine:
    for element in v:
        Graph_Initialization[vertex]=list(range(start,start+element))
        vertex+=1
        start=start+element
    Some_graph_of_twenty_vertices.append(Graph_Initialization)
    Graph_Initialization={0:list(range(1,11))}
    vertex=1
    start=11
```
